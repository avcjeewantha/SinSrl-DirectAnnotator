```
def replaceAtsWithComma(Resultslist): #replace @@@s With Commas
  res = []
  for i in Resultslist:
    if isinstance(i, list):
      insideL = []
      for j in i:
        x = j.replace('@@@', ',')
        x = x.strip()
        insideL.append(x)
      res.append(insideL)
  # if len(res) == 1:  
  #   return res[0]
  return res
```

```
def stringToList(ini_list): #Convert a string to a list
    # print(ini_list)
    res = ini_list.replace('],[','], [').replace('] ,[','], [').replace('] , [','], [').strip()[1:-1].split("], [")
    res = [[i] for i in res]
    # print(res)
    newRes = []
    for i in res:
      res = i[0].replace(', ,', ', @@@')  #to handle exceptions when "," is in the sentence
      newRes.append(res.strip('][').split(' , ')) 
    newRes = replaceAtsWithComma(newRes)
    # print(newRes)
    return newRes
```

```
def getMarks(s1 = x, s2 = y):
  predicates = getPredicateLists(s1, s2) #get predicates
  # print(predicates)
  noOfWordsInTheSentence = len(s1[0]) #get no. of words in the sentence
  # print(noOfWordsInTheSentence)
  noOfPredicatesInTheSentence = len(predicates[0]) #no. of predicates in the sentence
  # print(noOfPredicatesInTheSentence)
  predVerbs = removeSenses(predicates)
  # print(predVerbs)
  indexOfPredicatesInSentence = getIndexOfPredicateSentence(predVerbs[0], predVerbs[1]) #indexes of predicates
  # print(indexOfPredicatesInSentence)
  extraIndexes = indexOfPredicatesInSentence[1]
  marks = 0
  if noOfPredicatesInTheSentence > 0: #when there are predicates
    marksForSingleSentRelavantAPred = 100 / noOfPredicatesInTheSentence #marks for a tag list relevant to single predicate
    for k in range(len(indexOfPredicatesInSentence[0])): #iterate through indexes (mapping to s1-pred with s2 preds)
      predTags = [] #all the predicate tags in the sentence relevant to single predicate
      for i in range(noOfWordsInTheSentence): #iterate through words in the sentence
        # print(s1)
        # print(s2)
        if '.' in s1[k][i][1]: 
          predTags.append(s1[k][i][1])
      if len(predTags) == 0:
        marksForSRLTag = marksForSingleSentRelavantAPred / (noOfWordsInTheSentence - len(predTags))
      elif noOfWordsInTheSentence > len(predTags): #if all the words in the sentence are predicates (for a single predicate)
        marksForPredVerbTag = (marksForSingleSentRelavantAPred * 0.45) / len(predTags)
        marksForPredSense = marksForSingleSentRelavantAPred * 0.05 / len(predTags)
        marksForSRLTag = marksForSingleSentRelavantAPred * 0.5 / (noOfWordsInTheSentence - len(predTags))
      elif noOfWordsInTheSentence == len(predTags):
        marksForPredVerbTag = (marksForSingleSentRelavantAPred *0.9) / len(predTags)
        marksForPredSense = (marksForSingleSentRelavantAPred * 0.1) / len(predTags)
      for i in range(noOfWordsInTheSentence):
        if indexOfPredicatesInSentence[0][k] != 'No':
          if ('.' in s1[k][i][1]) and ('.' in s2[indexOfPredicatesInSentence[0][k]][i][1]):
            if s1[k][i][1].split('.')[0] == s2[indexOfPredicatesInSentence[0][k]][i][1].split('.')[0]:
              marks += marksForPredVerbTag
              if s1[k][i][1].split('.')[1] == s2[indexOfPredicatesInSentence[0][k]][i][1].split('.')[1]:
                marks += marksForPredSense
          elif s1[k][i][1] == s2[indexOfPredicatesInSentence[0][k]][i][1]:
            marks += marksForSRLTag
            ##code for extra tags
  elif noOfPredicatesInTheSentence == 0:
    # print(s1)
    # print(s2)
    marksForSingleTag = 100 / noOfWordsInTheSentence
    for i in range(noOfWordsInTheSentence):
      if s1[0][i][1] == s2[0][i][1]:
        marks += marksForSingleTag
    if len(predicates[1]) != 0:
      marks -= 5 * len(predicates[1])
  #----------------------------------------------------handle extra predicates----------------------
  noindexList = [] #indexes of 'No' preds
  if len(indexOfPredicatesInSentence[0]) != 0:
    for i in range(len(indexOfPredicatesInSentence[0])):
      if indexOfPredicatesInSentence[0][i] == 'No':
        noindexList.append(i)
  noOfNoIndexes = len(noindexList)
  noOfExtraIndexes = len(extraIndexes)
  remainingExtraIndexes = len(extraIndexes)
  if noOfNoIndexes > 0 and noOfNoIndexes < noOfExtraIndexes: #when No of missing predicates are lesser than extra predicates
    noOfPredTags = 0
    noCheckedOfExtraIndexes = 0
    for i in range(noOfNoIndexes): #loop through each "No" predicate
      for j in range(noOfWordsInTheSentence): #loop through each word
        if '.' in s1[noindexList[i]][j][1]:
          noOfPredTags += 1
      if noOfWordsInTheSentence != noOfPredTags:
        M = (marksForSingleSentRelavantAPred * 0.5) / (noOfWordsInTheSentence - noOfPredTags)
        for j in range(noOfWordsInTheSentence):
          if s1[noindexList[i]][j][1] == s2[noCheckedOfExtraIndexes][j][1]:
            marks += M
      noCheckedOfExtraIndexes += 1
    marks -= 5 * (noOfExtraIndexes - noCheckedOfExtraIndexes)
  if noOfNoIndexes > 0 and noOfNoIndexes > noOfExtraIndexes: #when No of missing predicates are greater than extra predicates
    noOfPredTags = 0
    noCheckedOfExtraIndexes = 0
    for i in range(noOfExtraIndexes): #loop through each extra predicate
      for j in range(noOfWordsInTheSentence): #loop through each word
        if '.' in s1[noindexList[i]][j][1]:
          noOfPredTags += 1
      if noOfWordsInTheSentence != noOfPredTags:
        M = (marksForSingleSentRelavantAPred * 0.5) / (noOfWordsInTheSentence - noOfPredTags)
        for j in range(noOfWordsInTheSentence):
          if s1[noindexList[i]][j][1] == s2[noCheckedOfExtraIndexes][j][1]:
            marks += M
      noCheckedOfExtraIndexes += 1
  return marks
```
